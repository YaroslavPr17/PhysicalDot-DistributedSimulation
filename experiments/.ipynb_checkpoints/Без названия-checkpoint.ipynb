```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(30,20)})
sns.set (font_scale=3)
```

```python
stat = pd.read_csv('./experiments/bodies/exp_data.multi_thread_repeated_threading.txt')
stat
```

```python
print(stat.dtypes)
stat.n_threads = stat.n_threads.astype('category')
```

```python
sns.lineplot(
    stat,
    x='n_bodies',
    y='mean_comp_time',
    hue='n_threads',
    linewidth = 1.5
);
```

```python
stat_pivot = pd.pivot(stat, 'n_bodies', 'n_threads', 'mean_comp_time')
stat_pivot
```

```python
for n_th in stat_pivot.index:
    sns.lineplot(
        stat_pivot.loc[n_th, :]
    )
```
